```python
# Import the Pathlib and csv module 

from pathlib import Path
import csv

csv_file = Path('C:/Users/adria/OneDrive/Desktop/Fintech Bootcamp/Homework/python-homework/Pybank/budget_data.csv') 

# Initalize the matrics for net profit, number of months and average change in sales. 
net_profit = 0
month_count = 0
net_change_list = []
month_change = []

# Initialize min key value pair
min_value = 0
min_key = ""

# Initialize max key value pair
max_value = 0
max_key = ""

# Initialize list of records
records = []


# Open file as readable object
with open(csv_file, 'r') as csvfile:
    csv_reader = csv.reader(csvfile)
    
    # Logic to skip first line (headers)
    csv_header = next(csv_reader)
    
    # Append new header 'change'
    csv_header.append("Change")
    
    # Append csv_header into records
    records.append(csv_header)
    
    # Logic to define net_profit (Substracting from previous row)
    first_row = next(csv_reader)
    month_count = 1
    net_profit = net_profit + int(first_row[1])
    previous_net = int(first_row[1])
    
    # Loop to find net_change for each row  
    for row in csv_reader:
        net_profit += int(row[1])
        month_count += 1 
        
        net_change = int(row[1])-previous_net
        previous_net = int(row[1])
        
        net_change_list = net_change_list + [net_change]
        month_change =  month_count - 1
        
        # Append Net change into change
        row.append(net_change)
        records.append(row)
        
       
        

        
        # logic for Min Key Value Pair
        if min_value == 0:
             min_key = row[0]
             min_value = row[2]
        elif row[2] < min_value:
            min_key = row[0]
            min_value = row[2]
            
        # logic for Max Key Value Pair
        if row[2] > max_value:
            max_key = row[0]
            max_value = row[2]
            
    # Logic for average month change     
    average_month_change = round(sum(net_change_list)/month_change,2)


        
print("Financial Analysis")
print("--------------------------")
print(f"Total Months: {month_count}")
print(f"Total: ${net_profit}")
print(f"Average Change: {average_month_change}")
print(f"Greatest Increase in Profits: {max_key} (${max_value})")
print(f"Greatest Decrease in Profits: {min_key} (${min_value})")
```
